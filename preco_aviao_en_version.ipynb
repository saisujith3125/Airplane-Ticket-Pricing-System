## **Dataset:**

**1 - Model:** The model of the aircraft (e.g. Boeing 737, Airbus A320). Represents an important feature that influences price based on brand and performance capabilities.

**2 - Year of Production:** The year the aircraft was manufactured. Helps calculate the age of the aircraft, a significant factor in depreciation.

**3 - Number of Engines:** The total number of engines on the aircraft (1 for piston engines, 2 for others).

**4 - Engine Type:** The type of engine, such as Turbofan or Piston. Different engine types affect performance, fuel efficiency, and price.

**5 - Capacity:** The passenger capacity of the aircraft. Larger aircraft with greater capacity tend to have higher base prices.

**6 - Range in km:** The maximum range the aircraft can travel without refueling. Greater range increases operational flexibility and cost.

**7 - Fuel consumption (L/hour):** Average fuel consumption of the aircraft. Turbofan engines consume less than piston engines, influencing operating costs.

**8 - Maintenance cost per hour:** The average maintenance cost per hour in USD. Higher maintenance costs can reduce demand and affect prices.

**9 - Age:** The age of the aircraft, calculated as . Older aircraft tend to have lower prices due to depreciation. 2023 - Year of production.

**10 - Sales region:** The region where the aircraft is being sold (e.g. Asia, Europe, North America). Regional demand and currency fluctuations can affect prices.

**11 - Price in USD:** The final price of the aircraft, serving as a target **_variable_** for forecasting.

**Key Features**

**Realism:** Prices are dynamically calculated based on multiple factors including age, capacity, range, and a noise factor to simulate real-world variability.

**Diversity:** Covers different aircraft models, engine types, and regions to provide a rich dataset for diverse analytical needs.

**Applications:** Ideal for regression modeling, feature engineering, and pricing analysis in the aviation industry.

**Step 3 - DataFrame Correction (Cleaning, Organization and Modeling)**

---

```python
# Replace the values ​​in the Engine Type and Sales Region columns

df_airplane["Sales Region"] = df_airplane["Sales Region"].replace({
    "Afrika": "Africa",
    "Asya": "Asia",
    "Avrupa": "Europe",
    "Avustralya": "Australia",
    "Güney Amerika": "South America",
    "Kuzey Amerika": "North America"
})

display(df_airplane.head(5))

# Function to replace the values ​​of the Model, Engine Type and Sales Region columns from text
# to numbers and vice versa
def code():
    if (df_airplane["Model"].dtypes == object) and (df_airplane["Engine Type"].dtypes == object) and (df_airplane["Sales Region"].dtypes == object):
        # Encode from text to number
        df_airplane["Model"] = df_airplane["Model"].replace({
            "Airbus A320": 0,
            "Airbus A350": 1,
            "Boeing 737": 2,
            "Boeing 777": 3,           
            "Bombardier CRJ200": 4,  
            "Cessna 172": 5                          
        })
        
        df_airplane["Engine Type"] = df_airplane["Engine Type"].replace({
            "Piston": 0,
            "Turbofan": 1
        })
        
        df_airplane["Sales Region"] = df_airplane["Sales Region"].replace({
            "Africa": 0,
            "Asia": 1,
            "Europe": 2,
            "Australia": 3,
            "South America": 4,
            "North America": 5
        })
    else:
        # Copy from number to text
        df_airplane["Model"] = df_airplane["Model"].replace({
            0: "Airbus A320",
            1: "Airbus A350",
            2: "Boeing 737",
            3: "Boeing 777",           
            4: "Bombardier CRJ200",  
            5: "Cessna 172"                          
        })
        
        df_airplane["Engine Type"] = df_airplane["Engine Type"].replace({
            0: "Piston",
            1: "Turbofan"
        })
        
        df_airplane["Sales Region"] = df_airplane["Sales Region"].replace({
            0: "Africa",
            1: "Asia",
            2: "Europe",
            3: "Australia",
            4: "South America",
            5: "North America"
        })
```

Dataframe corrected and remodeled with values ​​in English.

```python
# Importing the required libraries
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Creating a new dataframe for the AI, from the original dataframe
df_airplane_ia = df_airplane

display(df_airplane_ia)
```

---
**Step 1 - Prepare the dataframe for Artificial Intelligence**
---

```python
# Encode columns to numeric type for AI

# Encode Model column
encoder_model = LabelEncoder()
df_airplane_ia["Model"] = encoder_model.fit_transform(df_airplane_ia["Model"])

# Encode Engine Type column
encoder_motor = LabelEncoder()
df_airplane_ia["Engine Type"] = encoder_motor.fit_transform(df_airplane_ia["Engine Type"])

# Encode Sales Region column
encoder_region = LabelEncoder()
df_airplane_ia["Sales Region"] = encoder_region.fit_transform(df_airplane_ia["Sales Region"])

df_airplane_ia.info()

display(df_airplane_ia)
```

```python
# Prepare the dataframe for training

# y is the column of the dataframe to be predicted
y = df_airplane_ia["Price in USD"]

# x are the columns of the dataframe that will be used for prediction
x = df_airplane_ia.drop(columns=["Price in USD"])

# Separate training data and test data
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=42)
```

---

**Step 2 - Train the Artificial Intelligence**

---

```python
# Create the model -> Predict Prices in USD
# Decision Tree -> RandonForest
# Nearest Neighbors -> KNN -> KNeighbors
# Linear Regression -> RLN -> LinearRegression

# Create the AI
decision_tree_model = RandomForestRegressor()
knn_model = KNeighborsRegressor()
rln_model = LinearRegression()

# Train the AI
decision_tree_model.fit(x_treino, y_treino)
knn_model.fit(x_treino, y_treino)
rln_model.fit(x_treino, y_treino)
```

---

**Step 3 - Choose the best model**

---

```python
# Testing the models

# Predict test values ​​using Linear Regression models
forecast_decision_tree = decision_tree_model.predict(x_teste)
knn_prediction = knn_model.predict(x_teste)
forecast_rln = rln_model.predict(x_teste)

# Calculate and display the Mean Absolute Error (MAE) of models
print(f'MAE Decision Tree: {mean_absolute_error(y_teste, forecast_decision_tree):,.2f}')
print(f'MAE KNN: {mean_absolute_error(y_teste, knn_prediction):,.2f}')
print(f'MAE Linear Regression: {mean_absolute_error(y_teste, forecast_rln):,.2f}')

# Calculate and display the Mean Squared Error (MSE) of models
print(f'MSE Decision Tree: {mean_squared_error(y_teste, forecast_decision_tree):,.2f}')
print(f'MSE KNN: {mean_squared_error(y_teste, knn_prediction):,.2f}')
print(f'MSE Linear Regression: {mean_squared_error(y_teste, forecast_rln):,.2f}')

# Calculate and display model accuracy
display(decision_tree_model.score(x_teste, y_teste) * 100)
display(knn_model.score(x_teste, y_teste) * 100)
display(rln_model.score(x_teste, y_teste) * 100)

```

The best model is the Decision Tree.
